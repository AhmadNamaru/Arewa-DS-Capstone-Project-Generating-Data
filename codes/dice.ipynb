# From the Dice Examples

1. A number raised to the third power is a cube. Plot the first five cubic numbers, and then plot the first 5,000 cubic numbers.

```python
import matplotlib.pyplot as plt

first_five_cubes = [x**3 for x in range(1, 6)]

first_5000_cubes = [x**3 for x in range(1, 5001)]

# Ploting the first five cubic numbers
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(range(1, 6), first_five_cubes, marker='o')
plt.title('First Five Cubic Numbers', fontsize=16)
plt.xlabel('Number', fontsize=12)
plt.ylabel('Cube', fontsize=12)

# Ploting the first 5,000 cubic numbers
plt.subplot(1, 2, 2)
plt.plot(range(1, 5001), first_5000_cubes, marker='.')
plt.title('First 5,000 Cubic Numbers' , fontsize=16)
plt.xlabel('Number', fontsize=12)
plt.ylabel('Cube', fontsize=12)

plt.tight_layout()
plt.show()
```

2. Applying a colormap to the above cubes plot

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(range(1, 6), first_five_cubes, c=first_five_cubes, cmap='viridis', marker='o')
plt.title('First Five Cubic Numbers with Colormap', fontsize=12)
plt.xlabel('Number', fontsize=10)
plt.ylabel('Cube', fontsize=10)

plt.subplot(1, 2, 2)
plt.scatter(range(1, 5001), first_5000_cubes, c=first_5000_cubes, cmap='viridis', marker='.')
plt.title('First 5,000 Cubic Numbers with Colormap', fontsize=12)
plt.xlabel('Number', fontsize=10)
plt.ylabel('Cube', fontsize=10)

plt.tight_layout()
plt.show()
```
